```python
# 이미지 확대
img1 = cv2.imread('./kukaRobot.jpg')

w = img.shape[1] #가로 사이즈
h = img.shape[0] #세로 사이즈
print(w)
print(h)

width = round(w*2)
height = round(h*2)

up = cv2.resize(img1,(width,height))
cv2.imwrite('./kukaRobotup.jpg',up)
cv2.imshow('SizeUp',up)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 축소
img = cv2.imread('./kukaRobot.jpg')

w = img.shape[1]
h = img.shape[0]
print(w)
print(h)

width_d = round(w/2)
height_d = round(h/2)

down = cv2.resize(img,(width_d,height_d))#축소
cv2.imwrite('./kukaRobotdown.jpg',down)
cv2.imshow("SizeDown",down)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#이미지 밝기 조절
img1 = cv2.imread('./kukaRobot.jpg')
img2 = img1 + 50 #이미지 밝게
img3 = img1 - 50 #이미지 어둡게
img4 = cv2.hconcat([img1,img2,img3])
cv2.imshow('image3',img4)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 회전
img1 = cv2.imread('./kukaRobot.jpg')
height,width, channel = img1.shape

matrix = cv2.getRotationMatrix2D((width / 2, height / 2),45, 0.6)
#45도 회전(왼쪽), 크기 0.6
dst = cv2.warpAffine(img1,matrix,(width,height))

cv2.imshow('rotate',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 대칭
img = cv2.flip(img1,0)

img = cv2.imread('./kukaRobot.jpg')
cv2.imshow('1.image flip',cv2.flip(img,0)) #이미지 뒤집기
cv2.imshow('2.image flip',cv2.flip(img,-1)) #상하좌우대칭
cv2.imshow('3.image flip',cv2.flip(img,1))#좌우대칭
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#이미지 자르기
img = cv2.imread('./kukaRobot.jpg')
img_s = img[50:250, 100:250].copy()

cv2.imshow("image",img)
cv2.imshow("slice",img_s)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 이미지 색상공간변환
img = cv2.imread('./kukaRobot.jpg')
img_c = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
height, width , channel = img.shape
height2 , width2 = img_c.shape

print(channel)
cv2.imshow('color_change',img_c)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
