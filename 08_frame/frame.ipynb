```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
img1 = cv2.imread('stone.JPG', cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('Softbank_Pepper.JPG', cv2.IMREAD_GRAYSCALE)

src1 = cv2.resize(img1,(320,300))
src2 = cv2.resize(img2,(320,300))
```

## 산술연산

```python
if src1 is None or src2 is None:
    print('Image load failed!')
    sys.exit()

dst1 = cv2.add(src1, src2, dtype=cv2.CV_8U) # 덧셈
dst2 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.0) # 평균
dst3 = cv2.subtract(src1, src2) # 뺄셈
dst4 = cv2.absdiff(src1, src2) # 차이연산

plt.subplot(231), plt.axis('off'), plt.imshow(src1, 'gray'), plt.title('src1')
plt.subplot(232), plt.axis('off'), plt.imshow(src2, 'gray'), plt.title('src2')
plt.subplot(233), plt.axis('off'), plt.imshow(dst1, 'gray'), plt.title('add')
plt.subplot(234), plt.axis('off'), plt.imshow(dst2, 'gray'), plt.title('addWeighted')
plt.subplot(235), plt.axis('off'), plt.imshow(dst3, 'gray'), plt.title('subtract')
plt.subplot(236), plt.axis('off'), plt.imshow(dst4, 'gray'), plt.title('absdiff')
plt.show()
```

## 논리연산

```python
and1 = cv2.bitwise_and(src1, src2, dst=None, mask=None)
or1 = cv2.bitwise_or(src1, src2, dst=None, mask=None)
xor1 = cv2.bitwise_xor(src1, src2, dst=None, mask=None)
not1= cv2.bitwise_not(src1, dst=None, mask=None)

plt.subplot(231), plt.axis('off'), plt.imshow(and1, 'gray'), plt.title('and1')
plt.subplot(232), plt.axis('off'), plt.imshow(or1, 'gray'), plt.title('or1')
plt.subplot(233), plt.axis('off'), plt.imshow(xor1, 'gray'), plt.title('xor1')
plt.subplot(234), plt.axis('off'), plt.imshow(not1, 'gray'), plt.title('not1')
```
